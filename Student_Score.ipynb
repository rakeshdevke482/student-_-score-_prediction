Importing libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```


Load Dataset

```python
df = pd.read_csv('/content/student_scores.csv')
```

```python
df.head()
```

Scatter Plot Between Scores & Hours

```python
plt.scatter(df['Hours'],df['Scores'])
plt.xlabel('Hours')
plt.ylabel('Scores')
```

```python
X = df.iloc[:,0:1]
y = df.iloc[:,-1]
```

```python
y
```

Train-Test Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

Model Import

```python
from sklearn.linear_model import LinearRegression
```

Model Initialization

```python
lr = LinearRegression()
```

Model Training/Fit

```python
lr.fit(X_train,y_train)
```

```python
X_test
```

```python
y_test
```

```python
lr.predict(X_test.iloc[0].values.reshape(1,1))
```

Graph Plotting for Best Fit Line

```python
plt.scatter(df['Hours'],df['Scores'])
plt.plot(X_train,lr.predict(X_train),color='red')
plt.xlabel('Hours')
plt.ylabel('Scores')
```

```python
m = lr.coef_
```

```python
b = lr.intercept_
```

Linear Regression Equation

```python
# y = mx + b

m * 8.58 + b
```
